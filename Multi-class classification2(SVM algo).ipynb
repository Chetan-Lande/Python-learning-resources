## Problem Statement
Perform svm(Support vector machine) on iris dataset to create a ml model , in model by providing input as parts
(petal-sepal length & width) and gets the output as finds the specie of iris.

## Support Vector Machine
It works on both regerssion and classification, but it is suitable for classification, when will get the max 
accuracy , i.e very important.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
i = load_iris()
```

```python
dir(i)
```

```python
print(i.DESCR)
```

```python
#split data
x = i.data
y = i.target
```

```python
#create model
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y)
```

```python
len(train_x),len(test_x)
```

```python
#SVM Algo
from sklearn.svm import SVC
model = SVC()
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)     #to increase score changes random state
```

```python
sl = float(input("Enter the sepal length : "))
sw = float(input("Enter the sepal width : "))
pl = float(input("Enter the petal length : "))
pw = float(input("Enter the petal width : "))
n = model.predict([[sl,sw,pl,pw]])
if n == 0:
    print("setosa")
elif n == 1:
    print("versicolor")
elif n == 2:
    print("verginica")
else :
    print("not found")
```
