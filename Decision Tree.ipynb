## Problem Statement
used dataset = Bmi_male_female csv
find out what is gender of person by using height & weight. Male & Female is categorical value, by height &
weight to find whether the person is male/female.

## Decision Tree
so it is binary-class classification,either male/female (0/1)
so here find which algo is used?
here we used decision tree algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bmi_male_female.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# split data
x = df[['Height','Weight']]
y = df.Gender
```

but data is not ready to use in model,because in (y) data is not in numeric form it is in categorical format
first converts it(using label encoding) apply here. In pandas we have function dummies , by this apply one-hot
encoding like pd.get_dummies(y)


```python
# One hot encoding
Y = pd.get_dummies(y)
```

```python
# create ml model
```

```python
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")
model = DecisionTreeClassifier()
```

```python
model.fit(x,Y)
```

```python
model.predict([[190,60]])
```

```python
h = int(input("Enter the height of person : "))
w = int(input("Enter the weight of person : "))
p = model.predict([[h,w]])
if p[0][0] == False:
    print("Person - Male")
elif p[0][0] == True :
    print("Person - Female")
else :
    print("Person - Other")
```

above we use OHE

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y1 = le.fit_transform(y)
```

```python
model1 = DecisionTreeClassifier()
model1.fit(x,y1)
```

```python
model1.predict([[86,64]])
```

here o means female & 1 means male

```python
hi = int(input("Enter height of person : "))
wi = int(input("Enter weight of person : "))
o = model1.predict([[hi,wi]])[0]
if o == 0 :
    print("Person - Female")
else :
    print("Person - Male")
```
