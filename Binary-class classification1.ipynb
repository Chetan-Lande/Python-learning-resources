```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance.csv")
df.head()
```

```python
df.info()
```

```python
#split data
x = df[['age']]
y = df.bought_insurance
```

```python
# check which ml algo used here by using data visualization
plt.title("Age vs Insurance ")
plt.scatter(x,y, color = "red")
plt.xlabel("Age")
plt.ylabel("Insurance")
plt.show()
```

So by visuakization here we apply classification
In classification have different different kind of algorithm , like to apply a logistic regression ,support 
vector m/c algo or KNN algo.

```python
# check which ml algo used here by using data visualization
plt.title("Age vs Insurance")
plt.scatter(x,y, color = "red")
plt.plot(x,y, color = "blue")
plt.xlabel("Age")
plt.ylabel("Insurance")
plt.show()
```

so here linear regression won't work.
but in case you see this type of pattern ('s' like shaped) you can apply here logistic regerssion
this 's' like curve. Logistic Regression have sigmoid function(logic function).

```python
# Create moddel
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
model = LogisticRegression()
model.fit(x,y)
```

```python
pred = model.predict(x)
```

```python
n = int(input("Enter a age of person : "))
o = (model.predict([[n]]))[0]
if o == 0 :
    print("Insurance: Not bought")
else :
    print("Insurance: bought")
```

```python
#compare original value & predicted value by model
pred == y
```
